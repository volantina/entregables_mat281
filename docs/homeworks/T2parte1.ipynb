**Visualización de los datos**

```python
#graficamos la cantidad de información para cada estado
fig, ax  = plt.subplots(figsize=(16, 8))
fig.suptitle('Datos por estado', size = 20)
districts["state"].value_counts().head(15).plot(kind = 'pie', autopct='%1.1f%%', figsize=(16, 8), startangle=0).legend();
```

A travez de este gráfico notamos que el estudio está mayormente enfocado en Utah, Illinois, e Indiana.

```python
plt.figure(figsize = (16,8))
sns.countplot(data=districts, y="state", order=districts['state'].value_counts().index[:10])
plt.title('Top 10 State Distribution', size=18)
plt.show()
```

Aquí podemos ver los 10 estados que participan en mayor parte en el estudio, además de qué tanto influencian los datos cada uno.

```python
fig, ax  = plt.subplots(figsize=(16, 8))
fig.suptitle('Locale Type Distribution', size = 20)
districts["locale"].value_counts().head(15).plot(kind = 'pie', autopct='%1.1f%%', figsize=(16, 8), startangle=0).legend();
plt.show()
```

Podemos observar que la mayor parte del estudio fue enfocada en zonas suburbanas.

## Aprendizaje y plataformas

Crearemos un dataframe que relacione los sectores de educación y sus respectivos ip, con el porcentaje de instituciones que proveen ese sistema de educación.

```python
df=products.groupby('sector(s)').count()[['lp id']]
df['percent'] = (df['lp id']/df['lp id'].sum() *100)
x=list(df['percent'])
y=[]
for i in x:
    y.append(str(i))
df
```

```python
plt.figure(figsize = (16,8))
plt.title("Products Sector Distribution", size=20)
plt.ylabel('Sector(s)')
plt.xlabel('No. of providers')
sns.barplot(x=list(df['lp id']),y=df.index)
plt.show()
```

Notamos que el sistema PreK-12 predomina en la mayoría de los distritos estudiados, siguiendole el sistema PreK-12; Higher Ed; Corporate y luego PreK-12; Higher Ed. En general es un caso aislado que una institución ofrezca Higher Ed; Corporate o solo Corporate. Es decir, en la mayoría de los casos, las instituciones educativas se enfocan más en la educación parvularia y básica.

```python
plt.figure(figsize = (16,8))
sns.countplot(data=products, y="provider/company name", order=products['provider/company name'].value_counts().index[:10])
plt.title('Top 10 Provider Platform with the Most Product in 2020', size=18)
plt.show()
```

Aquí se puede visualizar los 10 proveedores de plataformas de aprendizaje más utilizadas por los sistemas educativos. Notamos una fuerte preferencia hacia Google LLC, y vemos que la cantidad de sistemas educativos que utilizan las demás plataformas es aproximadamente constante y bajo comparado con la plataforma líder.

```python
plt.figure(figsize = (16,8))
sns.countplot(x=products["primary essential function"]);
plt.xticks(rotation=90);
```

Se puede visualizar aquí la función principal de cada una de las plataformas. Notamos que la mayoría tiene como función ser una plataforma de aprendizaje digital, siguiendole las plataformas de recursos y referencias. Se podría inferir de esto que el cliente principal de las plataformas son los profesores parvularios y de educación básica (que utilizan en su mayoría las plataformas de aprendizaje digital), y estudiantes de educación media y superior (que utilizan en su mayoría plataformas de recursos y referencias).
