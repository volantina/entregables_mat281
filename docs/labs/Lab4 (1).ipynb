1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
#1: Scatter Plot
fig = plt.figure(figsize=(10, 5)) # ventana
plt.scatter(
    df['month_number'], # eje x
    df['total_profit'], # eje y
)
plt.title("Total profit scatter plot")
plt.xlabel("Mes")
plt.ylabel("Total profit")
#plt.legend(loc='upper left')
plt.show()
```

```python
#1: Line plot
fig = plt.figure(figsize=(10, 5)) # ventana
plt.plot(
    df['month_number'], # eje x
    df['total_profit'], # eje y
)
plt.title("Total profit simple line plot")
plt.xlabel("Mes")
plt.ylabel("Total profit")
#plt.legend(loc='upper left')
plt.show()
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
# grafico compueso

# grafico
# tamano del grafico
fig = plt.figure(figsize=(10, 5))

# graficar con for
colors = ["blue", "green","red","cyan", "magenta", "black","white"]
for i in range(1,6):
  plt.plot(
    df['month_number'], # eje x
    df[df.columns[i]], # eje y
    label=df.columns[i], # etiquetado
    color=colors[i], # color
    linewidth=1, # tamano de la curva

  )

plt.legend() # agregar etiquetado
plt.title("Venta de productos") # agregar titulo
plt.xlabel('Mes') # nombre eje x
plt.ylabel('Profits') # nombre eje y
plt.grid() # agregar grillado
plt.show() # mostrar grafico
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
# datos
# grafico facecream
fig = plt.figure(figsize=(10, 5))
plt.bar(
    df['month_number'], # eje x
    df['facecream'], # eje y
    # error mostrado en eje y
    align='center', # centrar nombre eje x
    color="blue", # color
    alpha=0.6 # intensidad del color
)
plt.title('Bar plot for facecream')
plt.xticks(df['month_number'])
plt.xlabel('Mes')
plt.ylabel('Profits')
plt.show()

# grafico facewash
fig = plt.figure(figsize=(10, 5))
plt.bar(
    df['month_number'], # eje x
    df['facewash'], # eje y
    # error mostrado en eje y
    align='center', # centrar nombre eje x
    color="blue", # color
    alpha=0.6 # intensidad del color
)
plt.title('Bar plot for facewash')
plt.xticks(df['month_number'])
plt.xlabel('Mes')
plt.ylabel('Profits')
plt.show()
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
dataa = df.drop(['month_number', 'total_units','total_profit'], axis = 1)
# New boxplot using stats_df
plt.figure(figsize=(10, 6))
sns.boxplot(data=dataa)
plt.show()
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
sizes = []
for columna in dataa.columns:
  sizes.append(dataa[columna].sum())
fig1, ax1 = plt.subplots(figsize=(8, 8))
ax1.pie(
    sizes,

    labels=dataa.columns,

    autopct='%1.1f%%',

    startangle=90
)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Pie chart for total profits')
plt.show()
```
